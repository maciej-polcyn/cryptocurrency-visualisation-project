```python
import pandas as pd

from sqlalchemy import create_engine

server = 'DESKTOP-U2JV8HI\\SQLEXPRESS'  # Note the double backslash for escaping
database = 'cryptoUpdated'

# Construct the connection string
connection_string = f'DRIVER=SQL Server;SERVER={server};DATABASE={database};Trusted_Connection=yes;'

# Create a SQLAlchemy engine using the connection string
engine = create_engine(f'mssql+pyodbc:///?odbc_connect={connection_string}')

# SQL queries to fetch data
query_curr = "SELECT * FROM cryptoUpdated.dbo.currencies;"
query_date = "SELECT * FROM cryptoUpdated.dbo.dateRange;"
query_market = "SELECT * FROM cryptoUpdated.dbo.marketPrices;"

# Use SQLAlchemy to execute the queries and load data into DataFrames
curr = pd.read_sql(query_curr, engine)
date = pd.read_sql(query_date, engine)
market = pd.read_sql(query_market, engine)

# Select values for calculation of correlation
second_value = input('Enter the symbol of a currency you want to compare: ')

btc = market[market['currency'] == 'BTC'][['date','price']]
second = market[market['currency'] == second_value][['date','price']]

# Merge data into one DataFrame and calculate correlation
merged = second.merge(btc, how='inner', on='date', suffixes=(f'_{second_value}', '_BTC'))
# merged[[f'price_{second_value}','price_BTC']].corr()
correlation = merged[f'price_{second_value}'].corr(merged['price_BTC'])

# Print results
if correlation >= 0.7:
    interpret = "a strong positive relationship."
elif correlation >= 0.3:
    interpret = "a moderate positive relationship."
elif correlation <= -0.3:
    interpret = "a moderate negative relationship."
elif correlation <= -0.7:
    interpret = "a strong negative relationship."
else: 
    interpret = 'no relationship.'

print(f'The correlation between prices of BTC and {second_value} is equal to ', round(correlation, 2),"which means there's", interpret)
print('Date range: ', merged['date'].min(), merged['date'].max())
```
